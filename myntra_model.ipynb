```python
import matplotlib.pyplot as plt 
import pandas as pd

# Uploading data 
df = pd.read_csv('C:\\Users\\Akshita Sharma\\Downloads\\myntra_dataset_ByScraping.csv')

# Cleaning data
df = df.dropna(subset=['brand_name','price','MRP','discount_percent','ratings','number_of_ratings'])

# Group brands by average ratings
brand_ratings = df.groupby('brand_name')['ratings'].mean().sort_values(ascending=False).head(20)

# Plot bar chart
plt.barh(brand_ratings.index, brand_ratings.values, color='blue', edgecolor='black')
plt.xlabel('Average Ratings')
plt.ylabel('Brand Name')
plt.title('BRANDS WITH HIGHEST RATING')
plt.gca().invert_yaxis()  # so the highest rating brand appears at the top
plt.show()

# Group brands by average ratings
brand_ratings = df.groupby('brand_name')['ratings'].mean().sort_values(ascending=False).tail(20)

# Plot bar chart
plt.barh(brand_ratings.index, brand_ratings.values, color='blue', edgecolor='black')
plt.xlabel('Average Ratings')
plt.ylabel('Brand Name')
plt.title('BRANDS WITH LEAST RATING')
plt.gca().invert_yaxis()  # so the highest rating brand appears at the top
plt.show()

#which brand has the maximum sales?

fig, axes = plt.subplots(1, 2, figsize=(15, 10))

# The amount of discount given by the brands 
axes[0].bar(
    df['brand_name'].head(10).sort_values(),
    df['discount_percent'].head(10).sort_values(),
    color='red'
)
axes[0].set_xlabel("BRANDS")
axes[0].set_ylabel("DISCOUNT GIVEN")
axes[0].set_title("THE AMOUNT OF DISCOUNT GIVEN PER BRAND")

# Which brand has the highest sales?
data = df['brand_name'].value_counts().head(5).index
sizes = df['brand_name'].value_counts().head(5).values

axes[1].pie(
    sizes,
    labels=data,
    autopct='%1.1f%%',
    colors=['gold','skyblue','coral','lightgreen','pink','orange','purple','cyan','lightgrey','lightblue']
)
axes[1].set_title("MAX SALES")

plt.tight_layout()
plt.show()



```
